### pd.merge will merge 2 dataframes. the 'inner' is a type of merge in which only if both dataframes have the value, the value will be added. 'on="Cust_id"' says that it will merge the 2 dataframes on the common column of "Cust_id"

```python
pd.merge(market_df,orders_df,how='inner',on='Ord_id')
```

```python
pd.merge(market_df,product_df,how='inner',on='Prod_id')
```

```python
pd.merge(market_df,shipping_df,how='inner',on='Ship_id')
```

```python
pd.merge(orders_df,shipping_df,how='inner',on='Order_ID')
```

<b>There are 4 types of merges:-
1. inner- it displays all common values
2. outer- it displays all values and replaces not found values with null
3. right- it displays all values in the second database given and replaces values not found in the first one with null
4. left- it displays all values in the first database given and replaces values not found in the second one with null
</b>

```python
df1 = pd.DataFrame({'Name': ['Aman', 'Joy', 'Rashmi', 'Saif'],
                    'Age': ['34', '31', '22', '33'],
                    'Gender': ['M', 'M', 'F', 'M']})

df2 = pd.DataFrame({'Name': ['Akhil', 'Asha', 'Preeti'],
                    'Age': ['31', '22', '23'],
                    'Gender': ['M', 'F', 'F']})

df1
```

```python
df2
```

```python
pd.concat([df1,df2], axis= 0)
```

<b>pd.concat() combines dataframes. the axis= 0 says that it should be added at the bottom of the dataframe. axis= 1 will add to the side.</b>

```python
pd.concat([df1,df2], axis= 0,ignore_index= True)
```

<b>ignore_index= True ignores the old indexes and makes its own, starting from 0</b>

```python
df1.append(df2, ignore_index= True)
```

<b>.append() function does the same thing as pd.concat(). the only difference is that there is no axis option in .append()</b>

```python
pd.concat([df1,df3],axis = 1)
```

```python
IPL_2018 = pd.DataFrame({'IPL Team': ['CSK', 'SRH', 'KKR', 'RR', 'MI', 'RCB', 'KXIP', 'DD'],
                         'Matches Played': [16, 17, 16, 15, 14, 14, 14, 14],
                         'Matches Won': [11, 10, 9, 7, 6, 6, 6, 5]})
IPL_2018.set_index('IPL Team', inplace = True)
IPL_2018
```

```python
IPL_2017 = pd.DataFrame({'IPL Team': ['MI', 'RPS', 'KKR', 'SRH', 'KXIP', 'DD', 'GL', 'RCB'],
                         'Matches Played': [17, 16, 16, 15, 14, 14, 14, 14],
                         'Matches Won': [12, 10, 9, 8, 7, 6, 4, 3]})
IPL_2017.set_index('IPL Team', inplace = True)
IPL_2017
```

```python
record = IPL_2018.add(IPL_2017, fill_value = 0)
record
```

```python
record['Win_Percent'] = round(100*(record['Matches Won']/record['Matches Played']))
record
```

```python
record.sort_values(by = ['Matches Won','Win_Percent'], ascending = False)
```
