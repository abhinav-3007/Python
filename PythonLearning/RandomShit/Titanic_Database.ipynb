```python
plt.figure(1,figsize = (15,5))

plt.subplot(1,3,1)
df.Survived[df.Age<=18].value_counts().plot(kind= "bar")
plt.title("Senior Citizens")
plt.xticks(rotation = 0)

plt.subplot(1,3,2)
df.Survived[(df.Age<65) & (df.Age>18)].value_counts().plot(kind= "bar")
plt.title("Adults")
plt.xticks(rotation = 0)

plt.subplot(1,3,3)
df.Survived[df.Age<=18].value_counts().plot(kind= "bar")
plt.title("Children")
plt.xticks(rotation = 0)

plt.show()
```

from these charts we can infer that senior citizens and children had the highest survivors:dead ratio.

```python
plt.figure(2,figsize= (10,5))

plt.subplot(1,2,1)
df.Survived[(df.Fare<=32)].value_counts().plot(kind= "bar")
plt.title("Low Fare")
plt.xticks(rotation = 0)

plt.subplot(1,2,2)
df.Survived[(df.Fare>32)].value_counts().plot(kind= "bar")
plt.title("High Fare")
plt.xticks(rotation = 0)

plt.show()
```

from these we can infer that people who paid larger amounts had a significantly better chance of survival

```python
plt.figure(3,figsize=(10,15))

plt.subplot(3,2,1)
df.Survived[(df.Sex=="male") & (df.Fare>32)].value_counts(sort=False).plot(kind= "bar", color = ["r","g"])
plt.title("Rich Male")
plt.xticks(rotation = 0)

plt.subplot(3,2,2)
df.Survived[(df.Sex=="female") & (df.Fare<=32)].value_counts(sort=False).plot(kind= "bar", color = ["r","g"])
plt.title("Rich Female")
plt.xticks(rotation = 0)

plt.subplot(3,2,3)
df.Survived[(df.Sex=="male") & (df.Fare<=32) & (df.Fare>8)].value_counts(sort=False).plot(kind= "bar", color = ["r","g"])
plt.title("Medium Male")
plt.xticks(rotation = 0)

plt.subplot(3,2,4)
df.Survived[(df.Sex=="female") & (df.Fare<=32) & (df.Fare>8)].value_counts(sort=False).plot(kind= "bar", color = ["r","g"])
plt.title("Medium Female")
plt.xticks(rotation = 0)

plt.subplot(3,2,5)
df.Survived[(df.Sex=="male") & (df.Fare<=8)].value_counts(sort=False).plot(kind= "bar", color = ["r","g"])
plt.title("Poor Male")
plt.xticks(rotation = 0)

plt.subplot(3,2,6)
df.Survived[(df.Sex=="female") & (df.Fare<=8)].value_counts(sort=False).plot(kind= "bar", color = ["r","g"])
plt.title("Poor Female")
plt.xticks(rotation = 0)

plt.show()
```
